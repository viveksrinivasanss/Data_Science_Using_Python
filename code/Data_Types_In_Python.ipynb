# Python Data Types

### List

```python
cheeses = ['Cheddar', 'Edam', 'Gouda']
numbers = [17, 123]
empty = []
print(cheeses, numbers, empty)
```

### List With Mixed Types

```python
['spam', 2.0, 5, [10, 20]]
```

### Lists are Mutable

```python
numbers = [17, 123]
numbers[1] = 5
print(numbers)
```

### Traversing Lists

```python
for cheese in cheeses:
    print(cheese)
```

```python
for i in range(len(numbers)):
    print(numbers[i],numbers[i] * 2)
```

```python
for x in empty:
    print('This never happens.')
```

### List Operations

```python
a = [1, 2, 3]
b = [4, 5, 6]
c = a + b
print(c)
```

```python
[0] * 4
```

```python
[1, 2, 3]*3
```

```python
[3 * i for i in [1, 2, 3]]
```

### List Slicing

```python
t = ['a', 'b', 'c', 'd', 'e', 'f']
t[1:3]
```

```python
t[:4]
```

```python
t[3:]
```

```python
num = list(range(1,21))
num
```

```python
num[0:21:2]
```

```python
num[1:21:2]
```

### List Methods

```python
t = ['a', 'b', 'c']
t.append('d')
print(t)
```

```python
t1 = ['a', 'b', 'c']
t2 = ['d', 'e']
t1.extend(t2)
print(t1)
```

```python
t = ['d', 'c', 'e', 'b', 'a']
t.sort()
print(t)
```

```python
t = ['a', 'b', 'c']
x = t.pop(1)
print(x,t)
```

```python
t = ['a', 'b', 'c']
t.remove('b')
print(t)
```

```python
t = ['a', 'b', 'c', 'd', 'e', 'f']
del t[1:5]
print(t)
```

```python
nums = [3, 41, 12, 9, 74, 15]
len(nums)
```

```python
sum(nums)
```

### Dictionaries

```python
eng2sp = dict()
eng2sp
```

```python
eng2sp['one'] = 'uno'
eng2sp
```

```python
eng2sp = {'one': 'uno', 'two': 'dos', 'three': 'tres'}
eng2sp
```

```python
print(eng2sp['two'])
```

```python
print(eng2sp['four'])
```

```python
len(eng2sp)
```

```python
eng2sp.keys()
```

```python
eng2sp.values()
```

```python
'one' in eng2sp
```

```python
'uno' in eng2sp
```

```python
"uno" in list(eng2sp.values())
```

### Iteration In Dict

```python
for key in eng2sp:
    print(key, eng2sp[key])
```

```python
for key,val in eng2sp.items():
    print(key,val)
```
